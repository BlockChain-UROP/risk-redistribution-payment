```python
import csv
import glob
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-whitegrid')
import numpy as np
from pandas import Series, DataFrame
```

```python
from cmc import coinmarketcap
from datetime import datetime
```

```python
crypto = 'ethereum'
start_date, end_date = datetime(2017,1,1), datetime(2019,1,1)

df_ethereum = coinmarketcap.getDataFor(crypto, start_date, end_date, fields = ['Average'])
df_ethereum.reset_index(inplace=True)  
```

```python
df_ethereum.head(5)
```

```python
crypto = 'tether'
start_date, end_date = datetime(2017,1,1), datetime(2019,1,1)

df_tether = coinmarketcap.getDataFor(crypto, start_date, end_date, fields = ['Average'])
df_tether.reset_index(inplace=True)  
```

```python
df_tether.head(5)
```

```python
plt.figure(num=None, figsize=(18, 10), dpi=80, facecolor='w', edgecolor='k')
plt.rcParams.update({'font.size': 20})
plt.xlabel("Number of Merchants")
plt.ylabel("Latency (ms)")

plt.plot(df_ethereum[['Date']], df_ethereum[['ethereum']])
```
